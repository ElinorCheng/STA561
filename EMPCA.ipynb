```python
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
from sklearn.mixture import GaussianMixture

# Generate a dataset with 5 features
X, y = make_blobs(n_samples=1000, centers=5, n_features=5, random_state=42)

# Fit a Gaussian mixture model to the data
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(X)

# Use the EMPCA algorithm to estimate the principal components of the data
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(gmm.sample(10000)[0])

# Visualize the reduced dataset
import matplotlib.pyplot as plt
plt.scatter(X_reduced[:, 0], X_reduced[:, 1])
plt.show()
```
